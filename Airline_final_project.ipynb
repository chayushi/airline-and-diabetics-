Age_group vs satisfaction

```python
group = []
for i in df.index:
    if (df.loc[i, 'Age'] > 0) and (df.loc[i,'Age'] <= 12):
        group.append('Child')
    elif (df.loc[i,'Age'] > 12) and (df.loc[i,'Age'] <= 64):
        group.append('Young')
    else:
        group.append('Older')
df.insert(4,'Age_group',group)
```

```python
plt.figure(figsize = [6,5])
sns.countplot(x = 'Age_group',hue = 'satisfaction', data = df)
plt.title('Age vs Satisfaction')
```

```python
sns.countplot(x = 'Gender', hue = 'satisfaction', data = df)
plt.legend(fontsize =6, title='Satisfaction', loc='upper right', labels=['Neutral/Dissatisfied', 'Satisfied'])
```

```python
plt.figure(figsize=[15,8])
sns.scatterplot(data=df, x='Flight Distance', y='Arrival Delay in Minutes', hue='satisfaction')
plt.title('Flight Distance vs Arrival Delay by Satisfaction')
plt.show()
```

```python
sns.countplot(x='Cleanliness', hue='satisfaction', data=df, palette = ['red', 'green'])
plt.title('Satisfaction vs Cleanliness')
plt.xlabel('cleanliness')
plt.ylabel('Count')
plt.legend(fontsize =8 , title='Satisfaction', loc='upper left', labels=['Neutral/Dissatisfied', 'Satisfied'])
plt.tight_layout()
plt.show()
```

```python
a =df.groupby('Class')['Cleanliness'].sum().index
b =df.groupby('Class')['Cleanliness'].sum()
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
a1=ax1.barh(df.groupby('Class')['Cleanliness'].sum().index,df.groupby('Class')['Cleanliness'].sum(),color=['r','g','yellow'])
ax1.bar_label(a1,fmt='%.2f')
ax1.legend(df.groupby('Class')['Cleanliness'].sum().index)
ax2.pie(df.groupby('Class')['Cleanliness'].sum(),
       labels=df.groupby('Class')['Cleanliness'].sum().index,autopct='%.2f%%')
plt.title('Number of people in  different Class that are satisfied by the cleanliness')

plt.legend(fontsize= 8,title='Class', loc='upper left')
plt.show()
```
